```python
import numpy as np
import os
import cv2
data_root = '/database/daehyeon/all_age_faces/aligned_faces'
file_list = os.listdir(data_root)
for i,image in enumerate(file_list):
    if i == 0: 
        img = cv2.imread(os.path.join(data_root,image))
        img = cv2.resize(img,(224,224))
        img = np.expand_dims(img,axis=0)
        print(img.shape)
    else:
        img_ = cv2.imread(os.path.join(data_root,image))
        img_ = cv2.resize(img_,(224,224))
        img_ = np.expand_dims(img_,axis=0)
        img = np.concatenate((img, img_), axis=0)
    print(i,'/',len(file_list))
# print(img.shape)
# np.save(img,'/database/daehyeon/all_age_faces/aligned_faces/data.npy')
```
