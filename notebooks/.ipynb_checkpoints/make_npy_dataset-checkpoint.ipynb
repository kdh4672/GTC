```python
import numpy as np
import cv2
import os
data_root = '/database/daehyeon/Stanford40/JPEGImages'
file_list = os.listdir(data_root)
for item in file_list:
    if 'npy' in item:
        file_list.remove(item)
        print(item)
for i,image in enumerate(file_list):
    xml = image.split('.jpg')[0] + '.xml'
    f = open("/database/daehyeon/Stanford40/XMLAnnotations/"+xml, 'r')
    while True:
        line = f.readline()
        if not line: break
        if '<xmax>' in line:
            xmax = int(line.split('>')[1].split('<')[0])
        if '<xmin>' in line:
            xmin = int(line.split('>')[1].split('<')[0])
        if '<ymax>' in line:
            ymax = int(line.split('>')[1].split('<')[0])
        if '<ymin>' in line:
            ymin = int(line.split('>')[1].split('<')[0])
    img = cv2.imread(os.path.join(data_root,image))
    cropped_img = img[ymin:ymax,xmin:xmax]
    img = cv2.resize(cropped_img,(224,224))
    cv2.imwrite(os.path.join('/database/daehyeon/Stanford40/cropped_images',image),img)

# print(img.shape)
# np.save('/database/daehyeon/Stanford40/cropped_images/data.npy',img)
```
