```python
import os
import numpy as np
file_list = os.listdir('/database/daehyeon/bu101/images/')
list_ = file_list
ct = 0
data_root = '/database/daehyeon/bu101/images/'
folder_list = os.listdir(data_root) ## swing,golf,....
label_list = []
for i,folder_name in enumerate(folder_list):
    label = folder_list.index(folder_name)
    file_list = os.listdir(os.path.join(data_root,folder_name))
    for j,image in enumerate(file_list):
        full_path = os.path.join(data_root,folder_name,image)
        label = image.split('_')[0]
        label_list.append(list_.index(label))
        
label = np.array(label_list)
np.save('/database/daehyeon/mit/label.npy',label)
```

```python
import numpy as np
data_root = '/database/daehyeon/all_age_faces/aligned_faces'
file_list = os.listdir(data_root)
for i,image in enumerate(file_list):
    if i == 0: 
        img = cv2.imread(os.path.join(data_root,image))
        img = cv2.resize(img,(224,224))
        img = np.expand_dims(img,axis=0)
        print(img.shape)
    else:
        img_ = cv2.imread(os.path.join(data_root,image))
        img_ = cv2.resize(img_,(224,224))
        img_ = np.expand_dims(img_,axis=0)
        img = np.concatenate((img, img_), axis=0)
    print(i)
print(img.shape)
# np.save('/database/daehyeon/all_age_faces/aligned_faces/data.npy',img)
```
